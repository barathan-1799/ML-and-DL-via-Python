```python
## (1) Importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# To ensure that image can be displayed in cell
%matplotlib inline

# Allows us to render and display images directly in the notebook
from IPython.display import Image 
```

```python
## (2) Displaying image directly
Image(filename = "M3 CS.jpg")
```

```python
## (3) Reading images using OpenCV

# Syntax: cv2.imread(filename, flag)
# Flag = 0 to load image as a grayscale one
# Flag = 1 to load image as a color one
# Flag = -1 to include alpha channel
m3_img = cv2.imread("M3 CS.jpg", 0)
print(m3_img)
```

```python
## (4) Displaying image attributes

# Printing size of image (height x width)
print("(1) Image size: ", m3_img.shape)

# Printing data type of image
print("(2) Data type of image: ", m3_img.dtype)
```

```python
## (5) Displaying image using Matplotlib

# Matplotlib uses different color maps, hence why the colours are different
plt.imshow(m3_img)
plt.show()
```

```python
# Displaying a grayscale version of the image using Matplotlib
#plt.imshow(m3_img, cmap="gray")
#plt.show()
```

```python
# Note: OpenCV stores images in BGR format while Matplotlib expects image in RGB format
# Thus, for correct display, we need to reverse channels of the image
m3_img_channels_reversed = m3_img[:, :, ::-1]
plt.imshow(m3_img_channels_reversed)
plt.show()
```

```python
## (6) Splitting and merging colour channels
m3_img_2 = cv2.imread("M3 CS.jpg", cv2.IMREAD_COLOR)
img_height = m3_img_2.shape[0]
img_width = m3_img_2.shape[1]
print("Height and width of image: ", img_height, "x", img_width)

# Splitting the image into B, G, R components
b, g, r = cv2.split(m3_img_2)

# Showing the channels
plt.figure(figsize=[img_height, img_width])
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title("Red channel", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title("Green channel", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title("Blue channel", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);

# Merge the individual channels into a BGR image
img_merged = cv2.merge((b, g, r))

# Show the merged output
plt.subplot(144); plt.imshow(img_merged[:, :, ::-1]); plt.title("Merged Output", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);
plt.show()
```

```python
## (7) Converting image to different color spaces (e.g., BGR to RGB)
m3_img_2_rgb = cv2.cvtColor(m3_img_2, cv2.COLOR_BGR2RGB)
plt.imshow(m3_img_2_rgb)
plt.show()
```

```python
# Another color space is HSV (Hue, Saturation, Value)
m3_img_2_hsv = cv2.cvtColor(m3_img_2, cv2.COLOR_BGR2HSV)
h, s, v  = cv2.split(m3_img_2_hsv)

# Showing the channels
plt.figure(figsize=[img_width, img_height])
plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title("Hue channel", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title("Saturation channel", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title("Value channel", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);

# Merge the individual channels into a HSV image
img_merged = cv2.merge((h, s, v))
img_merged_rgb = cv2.cvtColor(img_merged, cv2.COLOR_HSV2RGB)

# Show the merged output
plt.subplot(144); plt.imshow(img_merged_rgb); plt.title("Merged Output", fontsize=80); plt.xticks(fontsize=80); plt.yticks(fontsize=80);
plt.show()
```

```python
# (8) Modifying colour channels
h_new = h + 10
m3_img_2_merged = cv2.merge((h_new, s, v))
m3_img_2_rgb = cv2.cvtColor(m3_img_2_merged, cv2.COLOR_HSV2RGB)

# Showing the channels
plt.imshow(m3_img_2_rgb)
plt.show()
```

```python
## (9) Saving image
cv2.imwrite("M3_CS_SAVED.jpg", m3_img_2_rgb)

# Reading back the saved image
Image(filename = "M3_CS_SAVED.jpg")
```
