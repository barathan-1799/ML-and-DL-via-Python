```python
### Way of annotating images (and video frames):
### - Using lines
### - Using circles
### - Using rectangles
### - Using text
```

```python
## (1) Importing dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from IPython.display import Image

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (100, 200)
```

```python
## (2) Loading the image
m3_img = cv2.imread("M3 CS.jpg", cv2.IMREAD_COLOR)
m3_img = cv2.cvtColor(m3_img, cv2.COLOR_BGR2RGB)
plt.imshow(m3_img)
plt.show()
```

```python
## (3) Drawing a line
m3_img_copy = m3_img.copy()
left_point_coord = (30, 100)
right_point_coord = (80, 101)
line_color = (0, 255, 255)

cv2.line(m3_img_copy, left_point_coord, right_point_coord, line_color, thickness = 5, lineType = cv2.LINE_AA);
plt.imshow(m3_img_copy)
plt.show()
```

```python
## (4) Drawing a circle
m3_img_copy = m3_img.copy()
center = (50, 100)
radius = 10
circle_color = (0, 255, 255)

cv2.circle(m3_img_copy, center, radius, circle_color, thickness = 2, lineType = cv2.LINE_AA);
plt.imshow(m3_img_copy)
plt.show()
```

```python
## (5) Drawing a rectangle
m3_img_copy = m3_img.copy()
upper_left_coord = (40, 97)
bottom_right_coord = (80, 107)
rectangle_color = (0, 255, 255)

cv2.rectangle(m3_img_copy, upper_left_coord, bottom_right_coord, rectangle_color, thickness = 2, lineType = cv2.LINE_AA);
plt.imshow(m3_img_copy)
plt.show()
```

```python
## (6) Adding a text
m3_img_copy = m3_img.copy()

text = "G80 BMW M3 CS"
fontScale = 1.2
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 255)
fontThickness = 2
start_point_x = 50
start_point_y = 150
start_coord = (start_point_x, start_point_y)

cv2.putText(m3_img_copy, text, start_coord, fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA)

plt.imshow(m3_img_copy)
plt.show()
```
