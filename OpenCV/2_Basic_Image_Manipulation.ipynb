```python
### Basic image manipulation includes:
### - Changing values of individual pixels
### - Cropping, resizing, and flipping images
```

```python
## (1) Importing dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image

%matplotlib inline
```

```python
## (2) Reading an image
m3_img = cv2.imread("M3 CS.jpg", 1)

## (3) Converting the color space of the image from BGR to RGB
m3_img = cv2.cvtColor(m3_img, cv2.COLOR_BGR2RGB)

## (4) Printing the matrix of the image (i.e., pixel values)
print(m3_img)
print("Shape of image: ", m3_img.shape)

## (5) Displaying the image via Matplotlib
plt.imshow(m3_img)
plt.show()
```

```python
## (6) Accessing individual pixels within the image
# Accessing the first pixel
print(m3_img[0,0,0])
```

```python
## (7) Modifying image pixels
m3_img_copy = m3_img
m3_img_copy[0,0,0] = 255
m3_img_copy[0,0,1] = 255
m3_img_copy[0,0,2] = 255

print(m3_img_copy)
plt.imshow(m3_img_copy)
plt.show()
```

```python
## (8) Cropping images
# Specifying cropped region
cropped_region = m3_img[40:100, 10:100]
plt.imshow(cropped_region)
plt.show()
```

```python
## (9) Resizing an image
# Method 1: Specifying fx and fy
# fx is scale factor along horizontal axis
# fy is scale factor along vertical axis
resized_cropped_region_2x = cv2.resize(cropped_region, None, fx=2, fy=2)
plt.imshow(resized_cropped_region_2x)
plt.show()
```

```python
# Method 2: Specifying output image size while maintaining aspect ratio
desired_width = 100
aspect_ratio = desired_width / cropped_region.shape[1]
desired_height = int(cropped_region.shape[0]*aspect_ratio)
dim = (desired_height, desired_width)
print("New dimensions of image: ", dim)

# dsize takes it as (width x height)
resized_cropped_region = cv2.resize(cropped_region, dsize=dim[::-1], interpolation = cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
plt.show()
```

```python
## (10) Flipping image
# 0 = Flipping along x-axis (vertical flip)
# 1 = Flipping along y-axis (horizontal flip)
# -1 = Flipping along both axes

m3_img_flipped_horz = cv2.flip(m3_img, 1)
m3_img_flipped_vert = cv2.flip(m3_img, 0)
m3_img_flipped_both = cv2.flip(m3_img, -1)

plt.figure(figsize = [100,200])
plt.subplot(141); plt.imshow(m3_img_flipped_horz); plt.title("Horizontal flip");
plt.subplot(142); plt.imshow(m3_img_flipped_vert); plt.title("Vertical flip");
plt.subplot(143); plt.imshow(m3_img_flipped_both); plt.title("Both flipped");
plt.subplot(144); plt.imshow(m3_img); plt.title("Original");
plt.show()
```
