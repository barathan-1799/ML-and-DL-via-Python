```python
# (1) Importing dependencies
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# (2) Load the video
source = "./Transportation.mp4"  # source = 0 for webcam
cap = cv2.VideoCapture(source)

# Print error if video stream or file cannot be opened
if (cap.isOpened() == False):
    print("Error opening video stream or file")
```

```python
# (3) Read and display one frame
ret, frame = cap.read()
plt.imshow(frame[...,::-1])
plt.show()
```

```python
# (4) Display video from file
from IPython.display import HTML
HTML("""
<video width=512 controls>
    <source src="Transportation.mp4" type="video/mp4">
</video>
""")
```

```python
## (5) Write video
# fourcc is 4-character code of codec used to compress frames

# Default resolutions of the frame are obtained
# Convert resolutions from float to integer
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
frame_dim = (frame_width, frame_height)
FPS = 10

# Define codec and create VideoWriter object
out_avi = cv2.VideoWriter('Transportation_clip.avi', cv2.VideoWriter_fourcc("M", "J", "P", "G"), FPS, frame_dim)
out_mp4 = cv2.VideoWriter('Transportation_clip.mp4', cv2.VideoWriter_fourcc(*'XVID'), FPS, frame_dim)
```

```python
## (6) Read frames from video and write to the previous VideoWriter objects
# Read until video is completed
while(cap.isOpened()):
    ret, frame = cap.read()

    if ret == True:
        out_avi.write(frame)
        out_mp4.write(frame)
    else:
        break
```

```python
# Once done, release the VideoCapture and VideoWriter objects
cap.release()
out_avi.release()
out_mp4.release()
```
